2. How does broadcasting work in NumPy?
  - In NumPy, we can perform mathematical operations on arrays of different shapes. It enables NumPy to efficiently apply operations element-wise without explicitly copying or reshaping data. Broadcasting applies specific rules to determine whether two arrays can be aligned for operations:
    - **Check Dimensions**: Ensure the arrays have the same number of dimensions or expandable dimensions.
    - **Dimension Padding**: If arrays have different numbers of dimensions, the smaller array is left-padded with ones.
    - **Shape Compatibility**: Two dimensions are compatible if:
        They are equal, or One of them is 1.

      - The NumPy package breaks down a task into multiple fragments and then processes all the fragments parallelly.
      - The NumPy package integrates C, C++, and Fortran codes in Python. These programming languages have very little execution time compared to Python.

````python
# 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {'A': [1,2,3,4,5], 'B': [5,4,3,2,1], 'C': [3,6,9,12,15]}
df = pd.DataFrame(data)
corr = df.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
````

````python
# 6. Create a DataFrame and add a new column based on an existing column.
import pandas as pd
df = pd.DataFrame({'Course': ['Python', 'java', 'javascript'],
                   'Starting Date': ['03/02/2025','07/02/2025','15/02/2025'],
                    'Cost': [15000, 25000, 10000]})
print(f'Before adding a new coloum:')
print(df)
print('After adding a new Coloum:')
df['Discount_Price'] = df.apply(lambda row: row.Cost -
                                  (row.Cost * 0.1), axis = 1)
print(df)
````

````python
# 7. Write a program to perform element-wise multiplication of two NumPy arrays.
import numpy as np

arr1 = np.array([1,2,3,4,5])
arr2 = np.array([6,7,8,9,10])

result = np.multiply(arr1, arr2)

print(f'Array 1: {arr1}')
print(f'Array 2: {arr2}')
print(f'The element-wise multiplication is: {result}')
````
