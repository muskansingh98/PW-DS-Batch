# **Theoretical**

1. What is a Support Vector Machine (SVM)?
  - Support Vector Machine (SVM) is a supervised machine learning algorithm used for classification and regression tasks. It tries to find the best boundary known as hyperplane that separates different classes in the data. It is useful when you want to do binary classification like spam vs. not spam or cat vs. dog.

      The main goal of SVM is to maximize the margin between the two classes. The larger the margin the better the model performs on new and unseen data.

2. What is the difference between Hard Margin and Soft Margin SVM?
  - **Hard Margin SVM**
      - **Goal:** To find a hyperplane that perfectly separates the data into different classes with the largest possible margin.
      - **Assumption:** Assumes that the data is linearly separable, meaning there exists a straight line (or a hyperplane in higher dimensions) that can completely separate the classes without any misclassifications.
  - **Soft Margin SVM**
      - **Goal:** To find a hyperplane that provides a good balance between maximizing the margin and minimizing the number of misclassifications.
      - **Assumption:** Does not assume that the data is perfectly linearly separable. It allows for some training data points to be within the margin or even on the wrong side of the hyperplane.

3. What is the mathematical intuition behind SVM?
  - The mathematical intuition behind Support Vector Machines (SVM) is based on finding the best possible hyperplane that separates different classes in your data.
      1. **Hyperplane:** A hyperplane is a decision boundary that separates these data points into different classes. For a 2D dataset, this hyperplane is a line; for a 3D dataset, it's a plane, and so on.
      2. **Optimal Hyperplane:** The goal of SVM is to find the "optimal" hyperplane among all possible hyperplanes that can separate the data. The optimal hyperplane is the one that maximizes the margin between the classes.
      3. **Margin:** The margin is the distance between the hyperplane and the nearest data points from each class. These nearest points are called "support vectors." A larger margin means the hyperplane is further away from the data points of either class, which generally leads to better generalization on unseen data.
      4. **Maximizing the Margin:** Mathematically, finding the optimal hyperplane involves solving an optimization problem. This problem aims to maximize the distance between the hyperplane and the support vectors, subject to the constraint that all data points are on the correct side of the hyperplane.

4. What is the role of Lagrange Multipliers in SVM?
  - The role of Lagrange Multipliers in SVM is not explicitly described. However, in the standard mathematical formulation of Support Vector Machines, Lagrange Multipliers are a crucial tool used to solve the constrained optimization problem that finds the optimal hyperplane.

5. What are Support Vectors in SVM?
  - Support Vectors is not explicitly written. However, in the context of SVMs, Support Vectors are the data points that lie closest to the decision boundary, also known as the hyperplane. These specific data points are crucial because they directly influence the position and orientation of the hyperplane, and thus the margin.

6. What is a Support Vector Classifier (SVC)?
  - A Support Vector Classifier (SVC) is not explicitly present. However, within the context of SVMs and the code provided in your notebook, an SVC is a type of Support Vector Machine specifically used for classification tasks. It aims to find the optimal hyperplane to separate different classes in the data. You can see its usage in the code examples for classification problems like the Iris and Wine datasets.

7. What is a Support Vector Regressor (SVR)?
  - A Support Vector Regressor (SVR) is used for regression tasks. It is a type of Support Vector Machine that aims to find a function that best maps the input features to continuous output values while allowing for a certain margin of error. You can see an example of its usage in your code for the housing dataset.

8. What is the Kernel Trick in SVM?
  - The kernel trick is a method used in SVMs to enable them to classify non-linear data using a linear classifier. By applying a kernel function, SVMs can implicitly map input data into a higher-dimensional space where a linear separator (hyperplane) can be used to divide the classes. This mapping is computationally efficient because it avoids the direct calculation of the coordinates in this higher space.

9. Compare Linear Kernel, Polynomial Kernel, and RBF Kernel.
  - **Linear Kernel:** No mapping is needed as the data is already assumed to be linearly separable.
  - **Polynomial Kernel:** Maps inputs into a polynomial feature space, enhancing the classifier's ability to capture interactions between features.
  - **Radial Basis Function (RBF) Kernel:** Also known as the Gaussian kernel, it is useful for capturing complex regions by considering the distance between points in the input space.

10. What is the effect of the C parameter in SVM?
  - The parameter C in Support Vector Machines (SVMs) with a linear kernel controls the trade-off between the margin of the decision boundary and the accuracy of classifying the training data. Essentially, C determines how much the SVM should penalize misclassifications, influencing the complexity and accuracy of the model. the effect of different C values:
      - **Small C:** A small C prioritizes a larger margin, even if it means some training errors. This can lead to better generalization on unseen data if the training data is noisy.
      - **Large C:** A large C prioritizes classifying all training data points correctly, even if it means a smaller margin. This can lead to overfitting if the training data is not representative of the real-world data.

11. What is the role of the Gamma parameter in RBF Kernel SVM?
  - The gamma parameter in Support Vector Machines (SVMs) is a crucial hyperparameter that significantly influences the model's performance, particularly when using non-linear kernels like the Radial Basis Function (RBF) kernel. Understanding and tuning this parameter is essential for building an effective SVM model. Impact of Gamma on Model Complexity is:
      - A small Gamma value means the kernel has a broad influence, and the decision boundary is less sensitive to individual data points.
      - A large Gamma value means the kernel has a narrow influence, and the decision boundary is highly influenced by individual data points, potentially leading to a more complex and possibly overfitted model.

13. What is Bayes’ Theorem?
  - Bayes' Theorem is a mathematical formula that helps determine the conditional probability of an event based on prior knowledge and new evidence.

      It adjusts probabilities when new information comes in and helps make better decisions in uncertain situations.

      **Formula**
      ```
      P(A/B) = P(B/A) * P(A) / P(B)
      ```

      **Where**
        - P(A) and P(B) are the probabilities of events A and B, also, P(B) is never equal to zero.
        - P(A|B) is the probability of event A when event B happens,
        - P(B|A) is the probability of event B when A happens.

14. Explain the differences between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes.
  - **Gaussian Naïve Bayes:**
      - Feature Type: Handles continuous numerical features.
      - Assumption: Assumes that the continuous features within each class are distributed according to a Gaussian (Normal) distribution.
      - Probability Calculation: Estimates the class-conditional probability of a feature value based on the mean and variance of that feature in the training data for each class.
      - When to use: Use this when your features are measurements or other continuous values. An example from your notebook is using Gaussian Naïve Bayes on the Breast Cancer dataset (code example 25), which has continuous features.
  - **Multinomial Naïve Bayes:**
      - Feature Type: Suitable for discrete features, typically counts or frequencies. It's commonly used for text classification, where features represent the counts of words in a document.
      - Assumption: Assumes that the features follow a multinomial distribution.
      - Probability Calculation: Calculates the probability of observing a specific feature value given a class based on the relative frequency of that feature value in the training data for that class.
      - When to use: Use this for problems involving discrete counts, especially in text classification. Your notebook includes an example of Multinomial Naïve Bayes for text classification on the 20 Newsgroups dataset (code example 26).
  - **Bernoulli Naïve Bayes:**
      - Feature Type: Designed for binary features, where each feature indicates the presence or absence of a particular attribute.
      - Assumption: Assumes that the features follow a Bernoulli distribution.
      - Probability Calculation: Calculates the probability of a feature being present or absent given a class.
      - When to use: Use this for problems where features are binary (0 or 1). Your notebook demonstrates Bernoulli Naïve Bayes on a dataset with binary features (code example 28).

15. When should you use Gaussian Naïve Bayes over other variants?
  - You should use Gaussian Naïve Bayes over other variants when you have continuous numerical features.
      - **Gaussian Naïve Bayes:** This variant is designed for features that are continuous and are assumed to follow a Gaussian (normal) distribution within each class. It estimates the probability of a feature value belonging to a class based on the mean and standard deviation of that feature in the training data for that class.
      - **Multinomial Naïve Bayes:** This variant is suitable for discrete features, such as word counts or frequencies in text classification. It is based on the multinomial distribution.
      - **Bernoulli Naïve Bayes:** This variant is used for binary features, where each feature represents the presence or absence of something (e.g., whether a word appears in a document or not). It is based on the Bernoulli distribution.

16. What are the key assumptions made by Naïve Bayes?
  - The fundamental Naive Bayes assumption is that each feature makes an:
      - **Feature independence:** This means that when we are trying to classify something, we assume that each feature (or piece of information) in the data does not affect any other feature.
      - **Continuous features are normally distributed:** If a feature is continuous, then it is assumed to be normally distributed within each class.
      - **Discrete features have multinomial distributions:** If a feature is discrete, then it is assumed to have a multinomial distribution within each class.
      - **Features are equally important:** All features are assumed to contribute equally to the prediction of the class label.
      - **No missing data:** The data should not contain any missing values.

17. What are the advantages and disadvantages of Naïve Bayes?
  - **Advantages**
      - Easy to implement and computationally efficient.
      - Effective in cases with a large number of features.
      - Performs well even with limited training data.
      - It performs well in the presence of categorical features.
      - For numerical features data is assumed to come from normal distributions
  - **Disadvantages**
      - Assumes that features are independent, which may not always hold in real-world data.
      - Can be influenced by irrelevant attributes.
      - May assign zero probability to unseen events, leading to poor generalization.

18. Why is Naïve Bayes a good choice for text classification?
  - Naïve Bayes is considered a good choice for text classification for several reasons, stemming from its underlying principles and practical performance on textual data:
      - **Efficiency and Speed:** Naïve Bayes classifiers are computationally efficient and fast to train and predict, even with large datasets.
      - **Simplicity and Ease of Implementation:** Naïve Bayes is a relatively simple algorithm to understand and implement.
      - **Effective with High-Dimensional Data:** Text data is inherently high-dimensional due to the large vocabulary size. Naïve Bayes handles high-dimensional feature spaces well, even with the "naive" assumption of feature independence. While words are not perfectly independent, the assumption often holds reasonably well in the context of text, especially with appropriate text preprocessing.
      - **Works Well with Limited Training Data:** Naïve Bayes can perform well even with limited training data. This is beneficial in text classification scenarios where obtaining a large, labeled corpus can be challenging.
      - **Suitable for Categorical Features:** Text data, when represented using techniques like Bag-of-Words or TF-IDF, can be treated as having discrete or categorical features (counts or weights of words). Naïve Bayes models like Multinomial Naïve Bayes are well-suited for this type of data.
      - **Handling of Zero Probabilities (with Smoothing):** As discussed earlier, Laplace smoothing helps address the issue of zero probabilities for unseen words or word combinations in text, which is a common problem in text classification.

19. Compare SVM and Naïve Bayes for classification tasks.
  - **Choose SVM when:** You have a medium-sized dataset, want to capture complex feature interactions, and are willing to invest time in hyperparameter tuning for potentially higher accuracy.
  - **Choose Naïve Bayes when:** You have a large dataset, need a fast and easy-to-implement algorithm, or are working with text data where the feature independence assumption often holds reasonably well. It's also a good baseline model to start with.

````python
# 21. Write a Python program to train an SVM Classifier on the Iris dataset and evaluate accuracy.

iris = datasets.load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

clf = svm.SVC(kernel='linear')

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy of the SVM classifier on the Iris dataset: {accuracy}")
````

````python
# 22. Write a Python program to train two SVM classifiers with Linear and RBF kernels on the Wine dataset, then compare their accuracies.

wine = datasets.load_wine()
X = wine.data
y = wine.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

clf_linear = svm.SVC(kernel='linear')

clf_linear.fit(X_train, y_train)

y_pred_linear = clf_linear.predict(X_test)

accuracy_linear = accuracy_score(y_test, y_pred_linear)

clf_rbf = svm.SVC(kernel='rbf')

clf_rbf.fit(X_train, y_train)

y_pred_rbf = clf_rbf.predict(X_test)

accuracy_rbf = accuracy_score(y_test, y_pred_rbf)

print(f"Accuracy of the SVM classifier with Linear kernel: {accuracy_linear}")
print(f"Accuracy of the SVM classifier with RBF kernel: {accuracy_rbf}")

if accuracy_linear > accuracy_rbf:
    print("The Linear kernel performed better on this dataset.")
elif accuracy_rbf > accuracy_linear:
    print("The RBF kernel performed better on this dataset.")
else:
    print("Both kernels performed equally well on this dataset.")
````

````python
# 24. Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision boundary.

iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target

X = X[y != 2]
y = y[y != 2]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

clf = svm.SVC(kernel='poly', degree=3)

clf.fit(X_train, y_train)

h = .02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)

plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=plt.cm.coolwarm, edgecolor='k', s=20)
plt.title("SVM with Polynomial Kernel Decision Boundary")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
````

````python
# 26. Write a Python program to train a Multinomial Naïve Bayes classifier for text classification using the 20 Newsgroups dataset.

newsgroups_data = fetch_20newsgroups(subset='all', categories=['alt.atheism', 'soc.religion.christian', 'comp.graphics', 'sci.med'])

X = newsgroups_data.data
y = newsgroups_data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

vectorizer = CountVectorizer()
X_train_vectors = vectorizer.fit_transform(X_train)
X_test_vectors = vectorizer.transform(X_test)

mnb = MultinomialNB()
mnb.fit(X_train_vectors, y_train)

y_pred = mnb.predict(X_test_vectors)

accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy of the Multinomial Naïve Bayes classifier on the 20 Newsgroups dataset: {accuracy}")
````

````python
# 27. Write a Python program to train an SVM Classifier with different C values and compare the decision boundaries visually.

iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target

X = X[y != 2]
y = y[y != 2]

C_values = [0.1, 1, 10, 100]

h = .02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

plt.figure(figsize=(12, 8))
for i, C in enumerate(C_values):
    clf = svm.SVC(kernel='linear', C=C)
    clf.fit(X, y)

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.subplot(2, 2, i + 1)
    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
    plt.title(f'SVM with C = {C}')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')

plt.tight_layout()
plt.show()
````

````python
# 29. Write a Python program to apply feature scaling before training an SVM model and compare results with unscaled data.

iris = datasets.load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

svm_unscaled = SVC(kernel='linear')
svm_unscaled.fit(X_train, y_train)

y_pred_unscaled = svm_unscaled.predict(X_test)
accuracy_unscaled = accuracy_score(y_test, y_pred_unscaled)

print(f"Accuracy of SVM on unscaled data: {accuracy_unscaled}")

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

svm_scaled = SVC(kernel='linear')
svm_scaled.fit(X_train_scaled, y_train)

y_pred_scaled = svm_scaled.predict(X_test_scaled)
accuracy_scaled = accuracy_score(y_test, y_pred_scaled)

print(f"Accuracy of SVM on scaled data: {accuracy_scaled}")

if accuracy_scaled > accuracy_unscaled:
    print("\nAccuracy improved after feature scaling.")
elif accuracy_unscaled > accuracy_scaled:
    print("\nAccuracy decreased after feature scaling (unlikely for SVM, but possible).")
else:
    print("\nAccuracy is the same with and without feature scaling.")
````

````python
# 32. Write a Python program to train an SVM Classifier on an imbalanced dataset and apply class weighting and check it improve accuracy.

np.random.seed(42)
n_samples = 1000
X = np.random.rand(n_samples, 2)
y = (X[:, 0] + X[:, 1] > 1.0).astype(int)

X_majority = X[y == 1]
y_majority = y[y == 1]
X_minority = X[y == 0]
y_minority = y[y == 0]

n_minority = len(X_minority)
X_majority_downsampled, y_majority_downsampled = resample(
    X_majority, y_majority,
    replace=False,
    n_samples=int(n_minority * 0.5),
    random_state=42
)

X_imbalanced = np.vstack((X_majority_downsampled, X_minority))
y_imbalanced = np.hstack((y_majority_downsampled, y_minority))

shuffle_idx = np.random.permutation(len(X_imbalanced))
X_imbalanced = X_imbalanced[shuffle_idx]
y_imbalanced = y_imbalanced[shuffle_idx]

print("Imbalanced dataset created.")
print(f"Class distribution (0: {np.sum(y_imbalanced == 0)}, 1: {np.sum(y_imbalanced == 1)})")

X_train_imb, X_test_imb, y_train_imb, y_test_imb = train_test_split(
    X_imbalanced, y_imbalanced, test_size=0.3, random_state=42, stratify=y_imbalanced
)

svm_no_weight = SVC(kernel='linear')
svm_no_weight.fit(X_train_imb, y_train_imb)

y_pred_no_weight = svm_no_weight.predict(X_test_imb)
accuracy_no_weight = accuracy_score(y_test_imb, y_pred_no_weight)

print(f"\nAccuracy of SVM on imbalanced data (without class weighting): {accuracy_no_weight}")

svm_with_weight = SVC(kernel='linear', class_weight='balanced')

svm_with_weight.fit(X_train_imb, y_train_imb)

y_pred_with_weight = svm_with_weight.predict(X_test_imb)
accuracy_with_weight = accuracy_score(y_test_imb, y_pred_with_weight)

print(f"Accuracy of SVM on imbalanced data (with class weighting): {accuracy_with_weight}")

if accuracy_with_weight > accuracy_no_weight:
    print("\nAccuracy improved with class weighting.")
elif accuracy_no_weight > accuracy_with_weight:
    print("\nAccuracy decreased with class weighting.")
else:
    print("\nAccuracy is the same with and without class weighting.")
````

````python
# 38. Write a Python program to train an SVM Classifier using Stratified K-Fold Cross-Validation and compute the average accuracy.

iris = datasets.load_iris()
X = iris.data
y = iris.target

n_splits = 5

skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

accuracy_scores = []

for fold, (train_index, test_index) in enumerate(skf.split(X, y)):
    print(f"--- Fold {fold + 1}/{n_splits} ---")

    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    svm_clf = SVC(kernel='linear', random_state=42)
    svm_clf.fit(X_train, y_train)

    y_pred = svm_clf.predict(X_test)

    fold_accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(fold_accuracy)

    print(f"Accuracy for Fold {fold + 1}: {fold_accuracy}")

average_accuracy = np.mean(accuracy_scores)
std_accuracy = np.std(accuracy_scores)

print(f"\nAverage accuracy across {n_splits} folds: {average_accuracy}")
print(f"Standard deviation of accuracy across {n_splits} folds: {std_accuracy}")
````

````python
# 39. Write a Python program to train a Naïve Bayes classifier using different prior probabilities and compare performance.

breast_cancer = datasets.load_breast_cancer()
X = breast_cancer.data
y = breast_cancer.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

gnb_default_priors = GaussianNB()
gnb_default_priors.fit(X_train, y_train)

y_pred_default = gnb_default_priors.predict(X_test)
accuracy_default = accuracy_score(y_test, y_pred_default)

print(f"Accuracy of Naïve Bayes with default priors: {accuracy_default}")
print(f"Default learned priors: {gnb_default_priors.class_prior_}")

custom_priors = np.array([0.4, 0.6])

if np.sum(custom_priors) != 1.0:
    print("Warning: Custom priors do not sum to 1. Normalizing.")
    custom_priors = custom_priors / np.sum(custom_priors)
    print(f"Normalized custom priors: {custom_priors}")


gnb_custom_priors = GaussianNB(priors=custom_priors)
gnb_custom_priors.fit(X_train, y_train)

y_pred_custom = gnb_custom_priors.predict(X_test)
accuracy_custom = accuracy_score(y_test, y_pred_custom)

print(f"\nAccuracy of Naïve Bayes with custom priors ({custom_priors}): {accuracy_custom}")

if accuracy_custom > accuracy_default:
    print("\nAccuracy improved with custom prior probabilities.")
elif accuracy_default > accuracy_custom:
    print("\nAccuracy decreased with custom prior probabilities.")
else:
    print("\nAccuracy is the same with default and custom prior probabilities.")

another_custom_priors = np.array([0.7, 0.3])
gnb_another_custom_priors = GaussianNB(priors=another_custom_priors)
gnb_another_custom_priors.fit(X_train, y_train)
y_pred_another_custom = gnb_another_custom_priors.predict(X_test)
accuracy_another_custom = accuracy_score(y_test, y_pred_another_custom)
print(f"\nAccuracy of Naïve Bayes with custom priors ({another_custom_priors}): {accuracy_another_custom}")
````
