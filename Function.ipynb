6. Explain the concept of generators in Python and how they are defined.
  - In Python, a generator is a function that returns an iterator that produces a sequence of values when iterated over. Generators are useful when we want to produce a large sequence of values, but we don't want to store all of them in memory at once. We can define a generator function using the def keyword, but instead of the return statement we use the yield statement.

7. What are the advantages of using generators over regular functions?
  - There are following adventages of using generators over regular functions
    - **Memory Efficiency:** Does not store the entire sequence in memory, making it ideal for large or infinite sequences.
    - **Lazy Evaluation:** Generates values one at a time, pausing execution and resuming from where it left off.
    - **Maintain State:** Maintains its state between yields, allowing the function to pick up where it left off on subsequent calls.
    - **Convenience:** Generators simplify code for iterating over large datasets or streams of data without needing to manage state explicitly.

````python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_even(lst):
  total = 0
  for num in lst:
    if num % 2 == 0:
      total += num
  return total
a = sum_even([24,56,73,5,62])
print(f'Sum of all even number in list is {a}')
````

````python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def prime():
  start, end = 1,200
  print(f'Prime numbers between {start} and {end} are')
  for num in range(start, end+1):
    if num > 1:
      for i in range(2, num):
        if (num % i) == 0:
          break
      else:
        print(num, end=' ')
prime()
````

````python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.current_term = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_term < self.max_terms:
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            self.current_term += 1
            return result
        else:
            raise StopIteration

num_terms = int(input("Enter the term of number: "))
print(f'Fibonacci sequence up to {num_terms} terms :', end=" ")
fib_Sequence = FibonacciIterator(num_terms)
for num in fib_Sequence:
  print(num, end=" ")
````

````python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_2(exponent):
  for i in range(exponent+1):
    yield 2**i
exp = int(input("Enter the exponent: "))
print(f"The Power of 2 up to exponent {exp} are", end=" ")
for power in power_of_2(exp):
  print(power, end=" ")
````

````python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()

file_path = 'my_file.txt'
for line in read_file_line_by_line(file_path):
    print(line)
````

````python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
my_list = [('apple', 2), ('banana', 1), ('mango', 5), ('orange', 3), ('cherry', 4)]
sorted_list = sorted(my_list, key=lambda x: x[1])
print(f'Before sort a list of tuples {my_list}')
print(f'After sort a list of tuples {sorted_list}')
````

````python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
def celsius_to_fahrenheit(celsius):
  return (celsius * 9/5) + 32

cel_temp = [0, 10, 20, 30, 40]

fahr_temp = list(map(celsius_to_fahrenheit, cel_temp))
print(f'Celsius temperature: {cel_temp}')
print(f'Fahrenheit temperature: {fahr_temp}')
````

````python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(string):
  vowels = "aeiouAEIOU"
  filtered_characters = filter(lambda char: char not in vowels, string)
  return "".join(filtered_characters)

my_string = "Hello, World!"
string_without_vowels = remove_vowels(my_string)
print(f'Before removing vowel from a given string: "{my_string}"')
print(f'After removing vowel from a given string: "{string_without_vowels}"')
````
