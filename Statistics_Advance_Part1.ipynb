1. What is a random variable in probability theory?
  - Random variable is a fundamental concept in statistics that bridges the gap between theoretical probability and real-world data. A Random variable in statistics is a function that assigns a real value to an outcome in the sample space of a random experiment. For example: if you roll a die, you can assign a number to each possible outcome.
  
2. What are the types of random variables?
  - There are two basic types of random variables:
    - Discrete Random Variables (which take on specific values).
    - Continuous Random Variables (assume any value within a given range).

3. What is the difference between discrete and continuous distributions?
  - A continuous uniform distribution describing outcomes over a continuous range and discrete uniform distribution dealing with distinct, separate values.
  
4. What are probability distribution functions (PDF)?
  - A probability distribution function (PDF) is a function that describes the likelihood of a random variable taking on a particular value or set of values. For a continuous random variable, the PDF is used to calculate the probability of the variable falling within a specific range of values, while for a discrete random variable, it gives the probability of the variable taking on a specific value.

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
  - Cumulative Distribution Functions (CDF) give the probability that a random variable is less than or equal to a specific value. but probability distribution functions (PDF) describe the likelihood of a random variable falling within a small interval around a particular value.
  - CDF represents cumulative probabilities, showing how probabilities accumulate as you move along the variable's range. but PDF represents probabilities as a smooth curve, indicating how likely different outcomes are within a range of values.

6. What is a discrete uniform distribution?
  - Discrete uniform distribution is a probability distribution where a finite number of outcomes exist, and each outcome has an equal probability of occurring. In essence, it’s a discrete version of the continuous uniform distribution.

7. What are the key properties of a Bernoulli distribution?
  - The Bernoulli distribution is a discrete probability distribution that models the probability of a single trial with two possible outcomes: success or failure.

      **The key properties of a Bernoulli distribution is:**
      - **Probability Mass Function (PMF):** The probability mass function is possible outcome is:
          - P(X = 1) = p
          - P(X = 0) = 1 - p
          - where 'X' is the random variable.
      - **Expected Value (Mean):** The expected value or mean of a Bernoulli distribution is equal to the probability of success:
          - E(X) = p
      - **Variance:** The variance of a Bernoulli distribution is given by:
          - Var(X) = p(1 - p)
      - **Binary Outcomes:** The Bernoulli distribution only has two possible outcomes, often denoted as 1 (success) and 0 (failure).
      - Single Parameter: It is characterized by a single parameter, 'p', representing the probability of success. The probability of failure is (1 - p).

8. What is the binomial distribution, and how is it used in probability?
  - Binomial Distribution is a probability distribution used to model the number of successes in a fixed number of independent trials, where each trial has only two possible outcomes: success or failure. This distribution is useful for calculating the probability of a specific number of successes in scenarios like flipping coins, quality control, or survey predictions.

9. What is the Poisson distribution and where is it applied?
  - Probability distribution which is used to model the number of events that occur in a fixed interval of time or space is called Poisson distribution.

      The Poisson distribution has a wide range of applications in various fields:
        - **Telecommunications:** Modeling the number of calls received at a call center per hour.
        - **Insurance:** Predicting the number of accidents or claims filed within a specific time period.
        - **Customer Service:** Estimating the number of customers arriving at a store or website during a given time interval.
        - **Traffic Flow:** Analyzing the number of cars passing a particular point on a road in a certain amount of time.
        - **Defects in Manufacturing:** Determining the number of defects in a production line per unit of time or per unit of product.
        - **Natural Events:** Modeling the frequency of earthquakes, hurricanes, or other natural disasters within a region.
        - **Biology:** Studying the distribution of bacteria or cells in a sample.

10. What is a continuous uniform distribution?
  - Continuous uniform distribution is a probability distribution that assigns equal probability density to all outcomes within a specified interval. This means that any value within the interval has an equal chance of occurring. The probability density function (PDF) remains constant throughout the interval and is zero outside the interval.

11. What are the characteristics of a normal distribution?
  - The characteristics of a normal distribution are:
      - Normal Distribution Curve is symmetric about mean.
      - Normal Distribution is unimodal in nature, i.e., it has single peak value.
      - Normal Distribution Curve is always bell-shaped.
      - Mean, Mode, and Median for Normal Distribution is always same.
      - Normal Distribution follows Empirical Rule.

12. What is the standard normal distribution, and why is it important?
  - Standard normal distribution, also known as the z-distribution, is a special type of normal distribution. In this distribution, the mean (average) is 0 and the standard deviation (a measure of spread) is 1. This creates a bell-shaped curve that is symmetrical around the mean. It is important because of it's use for studying various Natural Phenomenon, studying of Financial Data, in Social Science for studying and predicting various parameters, etc.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
  - The Central Limit Theorem in Statistics states that as the sample size increases and its variance is finite, then the distribution of the sample mean approaches normal distribution irrespective of the shape of the population distribution.
  
      It is critical because it allows statisticians to make inferences about population parameters using sample statistics. It underpins many statistical methods, such as hypothesis testing and the creation of confidence intervals, by justifying the normality assumption of large sample statistics​.

14. How does the Central Limit Theorem relate to the normal distribution?
  - The Central Limit Theorem (CLT) essentially states that as you take larger and larger samples from any population (regardless of its shape), the distribution of the sample means will approach a normal distribution. The CLT links any population distribution to the normal distribution through the behavior of sample means. This is a powerful idea that allows us to apply the well-understood properties of the normal distribution to a wide range of statistical problems.

15. What is the application of Z statistics in hypothesis testing?
  - Z statistics play a crucial role in hypothesis testing when we want to compare a sample mean to a population mean, or to compare two sample means, under the assumption that the population variance is known or the sample size is large.

16. How do you calculate a Z-score, and what does it represent?
  - We use the following formula to calculate a z-score:
  ```
  z = (X – μ) / σ
  ```
  Where:
      - **z** is the z-score
      - **X** is a single raw data value
      - **μ** is the mean
      - **σ** is the standard deviation
  - It represents the standardized value of a data point relative to the mean and standard deviation of its distribution.
      - **Positive z-score:** The individual value is greater than the mean.
      - **Negative z-score:** The individual value is less than the mean.
      - **A z-score of 0:** The individual value is equal to the mean.

17. What are point estimates and interval estimates in statistics?
  - A point estimate provides a single, specific value as the best guess for a parameter, such as using the sample mean to estimate a population mean. In contrast, an interval estimate gives a range of values that is likely to contain the population parameter, often expressed as a confidence interval.

18. What is the significance of confidence intervals in statistical analysis?
  - Confidence intervals provide a measure of the uncertainty associated with statistical estimates, allowing for a more nuanced and informed interpretation of the results. They are an essential tool for quantifying uncertainty, expressing confidence, assessing precision, and making informed decisions based on statistical analysis.

19. What is the relationship between a Z-score and a confidence interval?
  - Z-scores are standardized values that represent the number of standard deviations a data point is from the mean of a standard normal distribution (mean = 0, standard deviation = 1).

20. How are Z-scores used to compare different distributions?
  - Z-scores allow for the comparison of data points from different distributions, allowing you to make meaningful comparisons even when the distributions have different scales, shapes, or units of measurement. An element having a z-score greater than 0 represents that the element is greater than the mean. An element having a z-score equal to 0 represents that the element is equal to the mean.

21. What are the assumptions for applying the Central Limit Theorem?
  - The CLT is a powerful tool, but it's important to be aware of the assumptions under which it holds true. Here are the key assumptions:
      - **Random Sampling:** Samples must be randomly selected from the population.
      - **Independence:** Samples must be independent of each other meaning that the outcome of one sample does not influence the outcome of another.
      - **Sufficiently large Sample Size:** Generally a sample size of 30 or more is considered large enough for it to work because the theorem says that with a sample size of 30 or more the sample mean tends to follow a normal distribution.
      - **Finite Mean and Variance:** Population data must have a finite mean and variance. This means that the data shouldn't have extreme values or be infinite.

22. What is the concept of expected value in a probability distribution?
  - In mathematics, the expected value (also known as the mean, expectation, or average) of a random variable is a measure of the central tendency or average outcome of that variable over many repetitions of an experiment.
  - The expected value is not necessarily the most likely outcome. It's the average outcome you would expect in the long run.

23. How does a probability distribution relate to the expected outcome of a random variable?
  - A probability distribution provides a comprehensive description of the likelihoods associated with each possible outcome of a random variable. The expected outcome, also known as the expected value, is a single number that summarizes the central tendency of the distribution.
