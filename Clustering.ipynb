# **Theoretical**

1. What is unsupervised learning in the context of machine learning?
  - Unsupervised learning is a type of machine learning where the algorithm learns from data that has not been labeled or classified. The goal is to find patterns, structures, or relationships within the data without prior knowledge of the desired output. It is often used for tasks like clustering, dimensionality reduction, and anomaly detection.

2. How does K-Means clustering algorithm work?
  - K-Means is an iterative algorithm that aims to partition data points into K distinct clusters. Here's a breakdown of the steps:
      1. **Initialization:** Choose the number of clusters, K. Then, randomly select K data points from your dataset as the initial cluster centroids.
      2. **Assignment:** Each data point is assigned to the nearest centroid based on a distance metric (commonly Euclidean distance). This creates K initial clusters.
      3. **Update:** Recalculate the centroids for each cluster. The new centroid is the mean of all data points assigned to that cluster.
      4. **Iteration:** Repeat steps 2 and 3 until the cluster assignments no longer change or a maximum number of iterations is reached. This means the centroids have stabilized.

3. Explain the concept of a dendrogram in hierarchical clustering.
  - A dendrogram in hierarchical clustering is a tree-like diagram that shows the sequence of mergers or divisions of clusters. The branches of the dendrogram represent the clusters, and the height of the merge point represents the distance or dissimilarity between the clusters.

4. What is the main difference between K-Means and Hierarchical Clustering?
  - The main difference between K-Means and Hierarchical Clustering lies in their approach to forming clusters:
      - **K-Means:** This is a partitioning method. You need to specify the number of clusters (K) beforehand. It iteratively assigns data points to the nearest cluster centroid and updates the centroids until they stabilize. K-Means aims to minimize the within-cluster variance.
      - **Hierarchical Clustering:** This is a tree-based method that doesn't require you to specify the number of clusters in advance. It builds a hierarchy of clusters, either by starting with individual data points and merging them (agglomerative) or by starting with all data points in one cluster and splitting them (divisive). The relationships between clusters are represented in a dendrogram, allowing you to choose the number of clusters by cutting the tree at a certain level.

5. What are the advantages of DBSCAN over K-Means?
  - DBSCAN (Density-Based Spatial Clustering of Applications with Noise) has several advantages over K-Means:
      - **Doesn't require specifying the number of clusters beforehand:** K-Means requires you to define the number of clusters (K) before running the algorithm, which can be challenging if you don't have prior knowledge about the data's structure. DBSCAN, on the other hand, determines the number of clusters based on the density of the data.
      - **Can find arbitrarily shaped clusters:** K-Means assumes that clusters are spherical and equally sized. DBSCAN can discover clusters of various shapes and sizes, making it more suitable for datasets with complex structures.
      - **Robust to outliers:** DBSCAN naturally identifies noise points (outliers) as data points that do not belong to any cluster. K-Means is sensitive to outliers, as they can significantly affect the cluster centroids.
      - **No need for empty clusters:** K-Means can sometimes result in empty clusters, which doesn't happen with DBSCAN.

6. When would you use Silhouette Score in clustering?
  - You would use the Silhouette Score when you want a quantitative measure to assess the quality of your clustering results, particularly when determining the appropriate number of clusters or comparing different clustering approaches.

7. What are the limitations of Hierarchical Clustering?
  - Hierarchical Clustering is a powerful technique, but it does have some limitations:
      - **Sensitivity to noise and outliers:** Outliers can significantly affect the clustering results, especially in agglomerative clustering, as they can influence the merging decisions at various levels of the hierarchy.
      - **Scalability:** For large datasets, Hierarchical Clustering can be computationally expensive.
      - **Difficulty in handling different densities and sizes of clusters:** Hierarchical clustering can sometimes struggle with clusters that have significantly different densities or sizes. The distance metrics used might not accurately reflect the structure of clusters in such cases.
      - **Irreversibility of decisions:** Once data points are merged into a cluster (in agglomerative clustering) or a cluster is split (in divisive clustering), these decisions are irreversible. This can lead to suboptimal results if an early decision was not ideal.
      - **Choosing the number of clusters:** While the dendrogram helps visualize the cluster hierarchy, determining the optimal number of clusters from the dendrogram can still be subjective and requires careful interpretation.
      - **Space complexity:** Storing the dissimilarity matrix (in agglomerative clustering) can require O(n^2) space, which can be a limitation for large datasets.

8. Why is feature scaling important in clustering algorithms like K-Means?
  - Feature scaling is crucial in clustering algorithms like K-Means, Hierarchical Clustering, and DBSCAN because these algorithms rely on distance calculations to group data points.

      Feature scaling is essential in distance-based clustering algorithms to ensure that all features contribute fairly to the clustering process, leading to more accurate, reliable, and interpretable results.

9. How does DBSCAN identify noise points?
  - DBSCAN identifies noise points based on the density of the data. It works:
      - **Core points:** A data point is considered a core point if there are at least min_samples (a parameter you set) data points within a radius of eps (another parameter you set) around it (including the point itself).
      - **Border points:** A data point is considered a border point if it is within the eps radius of a core point but is not a core point itself.
      - **Noise points:** A data point is considered a noise point if it is neither a core point nor a border point. This means that within its eps radius, there are fewer than min_samples data points, and it is not reachable from any core point.

10. Define inertia in the context of K-Means.
  - Inertia, also known as within-cluster sum of squares, is a metric used to measure the quality of clustering in K-means. It calculates the sum of squared distances between each data point and its assigned centroid. Lower inertia indicates better-defined clusters.

11. What is the elbow method in K-Means clustering?
  - In K-Means clustering, we start by randomly initializing k clusters and iteratively adjusting these clusters until they stabilize at an equilibrium point. However, before we can do this, we need to decide how many clusters (k) we should use.
      1. We iterate over a range of k values, typically from 1 to n (where n is a hyper-parameter you choose).
      2. For each k, we calculate the Within-Cluster Sum of Squares (WCSS).

12. Describe the concept of "density" in DBSCAN.
  - DBSCAN is a density-based clustering algorithm that groups data points that are closely packed together and marks outliers as noise based on their density in the feature space. It identifies clusters as dense regions in the data space separated by areas of lower density.

13. Can hierarchical clustering be used on categorical data?
  - Yes, hierarchical clustering can be used on categorical data, but it's not as straightforward as using it on numerical data, and you need to take some extra steps.

      Hierarchical clustering algorithms, like most clustering algorithms, rely on calculating distances or dissimilarities between data points to group them. Standard distance metrics like Euclidean distance (commonly used for numerical data) are not directly applicable to categorical data.

14. What does a negative Silhouette Score indicate?
  - A negative score [-1, 0] indicate that the samples might have got assigned to the wrong clusters. The Python Sklearn package supports the following different methods for evaluating Silhouette scores.

15. Explain the term "linkage criteria" in hierarchical clustering.
  - In hierarchical clustering, linkage criteria are used to determine the distance or similarity between clusters during the merging process. Linkage criteria play a crucial role in determining the structure and characteristics of the resulting clusters.
  
16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
  - K-Means' reliance on distance to centroids, its assumption of spherical and equally sized clusters, and its lack of a density-based approach make it less suitable for datasets where clusters have diverse sizes, shapes, or densities. In such cases, density-based algorithms like DBSCAN or hierarchical clustering with appropriate linkage criteria might be more effective.

17. What are the core parameters in DBSCAN, and how do they influence clustering?
  - The core parameters in DBSCAN are `eps` and `min_samples`.
      - **`eps` (epsilon):** This is the maximum distance between two samples for one to be considered as in the neighborhood of the other. It defines the radius around a data point. A smaller eps will result in more points being considered noise and potentially splitting larger clusters into smaller ones. A larger eps will connect more points, potentially merging separate clusters and including more noise points within clusters.
      - **`min_samples`:** This is the number of samples in a neighborhood for a point to be considered as a core point (including the point itself). A smaller min_samples means that sparser regions can be considered clusters, making the algorithm more sensitive to noise. A larger min_samples makes the algorithm more robust to noise and suitable for finding dense clusters.

18. How does K-Means++ improve upon standard K-Means initialization?
  - it improves upon standard K-Means initialization:
      - **Standard K-Means Initialization (Random):**
          1. Randomly selects K data points from the dataset as the initial centroids.
          2. Drawback: This can result in initial centroids that are very close to each other or all located within a single cluster, leading to poor clustering or requiring many iterations to converge.
      - **K-Means++ Initialization:**
          K-Means++ uses a probabilistic approach to select the initial centroids, aiming to spread them out across the data space.


19. What is agglomerative clustering?
  - The agglomerative clustering is the most common type of hierarchical clustering used to group objects in clusters based on their similarity. Itâ€™s also known as AGNES (Agglomerative Nesting). The algorithm starts by treating each object as a singleton cluster.

20. What makes Silhouette Score a better metric than just inertia for model evaluation?
  - While inertia is a good starting point for understanding within-cluster variance, the Silhouette Score provides a more nuanced and informative evaluation of clustering quality by considering both how compact the clusters are and how well-separated they are from each other. It's a more robust metric for choosing the number of clusters and comparing different clustering results.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import make_moons
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.datasets import make_circles
from sklearn.metrics import silhouette_samples
from sklearn.datasets import load_wine
from sklearn.datasets import make_blobs
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# 21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot.

X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.8, random_state=42)

kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
kmeans.fit(X)
kmeans_labels = kmeans.predict(X)

# Visualize using a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis', s=50)
plt.title("K-Means Clustering on Synthetic Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
# 22. Load the Iris dataset and use Agglomerative Clustering to group the data into 3 clusters. Display the first 10 predicted labels.

iris = load_iris()
X = iris.data

# Apply Agglomerative Clustering
agg_clustering = AgglomerativeClustering(n_clusters=3)
agg_labels = agg_clustering.fit_predict(X)

# Display the first 10 predicted labels
print("First 10 predicted labels:")
print(agg_labels[:10])
```

```python
# 23. Generate synthetic data using make_moons and apply DBSCAN. Highlight outliers in the plot.

X, y = make_moons(n_samples=200, noise=0.05, random_state=42)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.1, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)

# Highlight outliers in the plot
plt.figure(figsize=(8, 6))
# Plot non-outliers
plt.scatter(X[dbscan_labels != -1, 0], X[dbscan_labels != -1, 1], c=dbscan_labels[dbscan_labels != -1], cmap='viridis', s=50, label='Clusters')
# Plot outliers
plt.scatter(X[dbscan_labels == -1, 0], X[dbscan_labels == -1, 1], c='red', s=50, marker='x', label='Outliers')
plt.title("DBSCAN Clustering on Make Moons Data with Outliers Highlighted")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
# 24. Load the Wine dataset and apply K-Means clustering after standardizing the features. Print the size of each cluster.

wine = load_wine()
X = wine.data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10) # Wine dataset has 3 classes
kmeans.fit(X_scaled)

cluster_sizes = np.bincount(kmeans.labels_)
print("Size of each cluster:")
for i, size in enumerate(cluster_sizes):
    print(f"Cluster {i}: {size}")
```

```python
# 28. Load the Digits dataset, reduce it to 2D using PCA, and visualize clusters from K-Means.

digits = load_digits()
X = digits.data
y = digits.target

pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X)

kmeans = KMeans(n_clusters=10, random_state=42, n_init=10) # Digits dataset has 10 classes
kmeans.fit(X_pca)
kmeans_labels = kmeans.predict(X_pca)

# Visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans_labels, cmap='viridis', s=50)
plt.title("KMeans Clustering on PCA-reduced Digits Dataset (2D)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(label="Cluster")
plt.show()
```

```python
# 29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.

X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.8, random_state=42)

silhouette_scores = []

# Evaluate silhouette scores for k = 2 to 5
for k in range(2, 6):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    score = silhouette_score(X, kmeans.labels_)
    silhouette_scores.append(score)

# Display as a bar chart
plt.figure(figsize=(8, 6))
plt.bar(range(2, 6), silhouette_scores)
plt.title("Silhouette Scores for KMeans Clustering (k=2 to 5)")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.xticks(range(2, 6))
plt.show()
```

```python
# 30. Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage.

iris = load_iris()
X = iris.data
y = iris.target

linked = linkage(X, 'average')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked,
            orientation='top',
            labels=iris.target_names[y],
            distance_sort='descending',
            show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram (Average Linkage) on Iris Dataset')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python
# 31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

X, y = make_blobs(n_samples=300, centers=3, cluster_std=1.5, random_state=42)

# Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X)
kmeans_labels = kmeans.predict(X)

# Visualize with decision boundaries
plt.figure(figsize=(8, 6))
ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(
    kmeans,
    X,
    response_method="predict",
    cmap='viridis',
    alpha=0.8,
    ax=ax,
    xlabel="Feature 1",
    ylabel="Feature 2",
)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis', s=50, edgecolors='k')
plt.title("KMeans Clustering with Decision Boundaries on Overlapping Blobs")
plt.show()
```

```python
# 47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count.

X, y = make_blobs(n_samples=500, n_features=2, centers=3, cluster_std=0.6, random_state=42)

dbscan = DBSCAN(eps=0.5, min_samples=5)

dbscan_labels = dbscan.fit_predict(X)

noise_mask = (dbscan_labels == -1)
clustered_mask = (dbscan_labels != -1)

noise_points = X[noise_mask]
clustered_points = X[clustered_mask]

unique_cluster_labels = np.unique(dbscan_labels[clustered_mask])

num_clusters = len(unique_cluster_labels)
num_noise_points = len(noise_points)

print(f"Number of clusters identified: {num_clusters}")
print(f"Number of noise points identified: {num_noise_points}")
```
